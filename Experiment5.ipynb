```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = np.genfromtxt("lr_dataset.csv", delimiter=",", skip_header=True)
type(data)
```

```python
X = data[:,0].reshape(-1,1)
X.shape[0]
```

```python
ones = np.ones([X.shape[0], 1])
ones.shape
```

```python
X = np.concatenate([ones, X], 1)
```

```python
Y = data[:,1].reshape(-1,1)
```

```python
plt.scatter(data[:,0].reshape(-1,1), Y)
```

```python
alpha = 0.0001
iters = 10000
```

```python
theta = np.array([[1.0, 1.0]])
theta.shape
```

```python
def computeCost(X, Y, theta):
    inner = np.power((np.dot(X, theta.T) - Y), 2)
    return np.sum(inner) / (2*len(X))

computeCost(X, Y, theta)
```

```python
def gradientDescent(X, Y, theta, alpha, iters):
    for i in range(iters):
        theta = theta - (alpha/len(X) * np.sum((np.dot(X, theta.T) - Y) * X, axis=0))
        cost = computeCost(X, Y, theta)
    return theta, cost

g, cost = gradientDescent(X, Y, theta, alpha, iters)
print(g, cost)
```

```python
plt.scatter(data[:,0].reshape(-1,1), Y)
axes = plt.gca()

X_vals = np.array(axes.get_xlim())
Y_vals = g[0][0] + g[0][1]*X_vals
plt.plot(X_vals, Y_vals, "r--")
```
