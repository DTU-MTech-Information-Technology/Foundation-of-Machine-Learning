```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.stats import uniform, norm, expon, binom, poisson
```

```python
# plot discrete functions
def plot_distribution(dist, x, title):
    pdf = dist.pdf(x)
    cdf = dist.cdf(x)

    plt.figure(figsize=(8,4))

    plt.subplot(1, 2, 1)
    plt.plot(x, pdf, 'r-', lw=2)
    plt.title(f"PDF of {title}")
    plt.xlabel("x")
    plt.ylabel("Probability Density")

    plt.subplot(1, 2, 2)
    plt.plot(x, cdf, 'b-', lw=2)
    plt.title(f"CDF of {title}")
    plt.xlabel("x")
    plt.ylabel("Cumulative Probability")

    plt.tight_layout(pad=5.0)
    plt.show()
```

```python
# plot continous functions
def plot_discrete_distribution(dist, x, title):
    pmf = dist.pmf(x)
    cdf = dist.cdf(x)

    plt.figure(figsize=(8,4))

    plt.subplot(1, 2, 1)
    plt.bar(x, pmf, color="r")
    plt.title(f"PDF of {title}")
    plt.xlabel("Number of successes")
    plt.ylabel("Probability")

    plt.subplot(1, 2, 2)
    plt.step(x, cdf, color="b")
    plt.title(f"CDF of {title}")
    plt.xlabel("Number of successes")
    plt.ylabel("Cumulative Probability")

    plt.tight_layout(pad=5.0)
    plt.show()
```

```python
# Uniform Distribution
a, b = 0, 10 # start and end of the uniform distribution
x_uniform = np.linspace(a, b, 100)
uniform_dist = uniform(a, b-a)
plot_distribution(uniform_dist, x_uniform, "Uniform Distribution")
```

```python
# Normal Distribution
mean, std = 0, 1 # mean and standard deviation
x_norm = np.linspace(-5, 5, 100)
normal_dist = norm(mean, std)
plot_distribution(normal_dist, x_norm, "Normal Distribution")
```

```python
# Exponential Distribution
scale = 1 # 1/lambda
x_exp = np.linspace(0, 10, 100)
expon_dist = expon(scale=scale)
plot_distribution(expon, x_exp, "Exponential Distribution")
```

```python
# Binomial Distribution
n, p = 10, 0.5 # no. of trials and probability of success
x_binom = np.arange(0, n+1)
binom_dist = binom(n, p)
plot_discrete_distribution(binom_dist, x_binom, "Binomial Distribution")
```

```python
# Poisson Distribution
lambda_poisson = 4 # mean no. of events
x_poisson = np.arange(0, 15)
poisson_dist = poisson(lambda_poisson)
plot_discrete_distribution(poisson_dist, x_poisson, "Poisson Distribution")
```
